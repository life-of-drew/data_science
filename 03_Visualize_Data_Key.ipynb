# How to Visualize Data in Python

## Learning Objectives
When exploring data, one of the most important things we can do is visualize it so we can better understand it. Like the popular saying "*A picture is worth a thousand words*‚Äù, visualizations are sometimes more useful than summary statistics in helping us understand our data. This is because visualizations are a great tool for asking and answering questions about data. Depending on the type of question we are trying to answer, there are four major types of visualizations we could use. They are relationship, distribution, comparison and composition. By the end of this tutorial, you will have learned:

+ how to create a relationship visualization
+ how to create a distribution visualization
+ how to create a comparison visualization
+ how to create a composition visualization

Let's import and preview the data we will use to illustrate how to visualize data.

```python
import pandas as pd
vehicles = pd.read_csv("vehicles.csv")
vehicles.head()
```

## How to create a Comparision Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a countplot of the 'class' column
sns.countplot(data=vehicles, y='class', hue='class', palette='pastel',legend=False)
plt.title('Count of Vehicles by Class')
plt.xlabel('Distribution')
plt.ylabel('Class')
plt.show()
```

```python
plt.figure(figsize=(12, 6))  # Width=12, Height=6 inches
sns.boxplot(x='class', y='co2emissions', data=vehicles)
plt.title('CO2 Emissions by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('CO2 Emissions')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

**Interpretation**

**Highest Median Emissions:**
Vans have higher median CO2 emissions, suggesting these classes generally emit more CO2.

**Lower Emission Classes:**
Compact Cars, Midsize Cars, Subcompact Cars show lower medians, indicating they're more fuel-efficient or emit less CO2 on average.

## How to create a Relationship Visualization

```python
# Insight: Vehicles with higher mileage ratings emit less carbon

#figsize=(8, 5) makes the plot more readable.
#Improve readability by adjusting point transparency with alpha if points overlap:

vehicles.plot(kind='scatter', x='citympg', y='co2emissions', figsize=(8, 5), alpha=0.6)
plt.title('CO2 Emissions vs City Miles per Gallon')
plt.xlabel('City Miles per Gallon')
plt.ylabel('CO2 Emissions');
```

## How to create a Distribution Visualization

```python
# Insight: Values from the dataset are concentrated in 300 to 700 grams per mile range
vehicles['co2emissions'].plot(kind = 'hist')

plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions');
```

## How to create a Composition Visualization

```python
# Create a crosstab of class vs transmission type
stack_data_by_transmission = pd.crosstab(vehicles['class'], vehicles['transmissiontype'])

# Plot as a stacked bar chart
stack_data_by_transmission.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 6),
    colormap='coolwarm'
)

# Add chart elements
plt.title('Vehicle Count by Class and Transmission Type')
plt.xlabel('Vehicle Class')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=90, ha='right')

#It uses a coordinate system where (0, 0) is the bottom-left of the axes, and (1, 1) is the top-right.
plt.legend(title='Transmission Type', bbox_to_anchor=(1,1), loc='upper left')
plt.show()
```
